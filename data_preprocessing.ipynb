```python
import os

data_dir = 'data'

def list_subdirectories(directory):
    return [item for item in os.listdir(directory)
            if os.path.isdir(os.path.join(directory, item))]


ids = list_subdirectories(data_dir)
ids
```

```python
int_ids = sorted([int(item) for item in ids])
int_ids
```

```python
print('업데이트된 데이터 개수', len(int_ids))
print('테이블의 데이터 개수', max(int_ids))
```

```python
import pandas as pd

meta = pd.read_csv('meta.csv')
meta
```

```python
# 결측 데이터 제거
processed_meta = meta[meta['index'].isin(int_ids)]
processed_meta
```

```python
processed_meta.to_csv('processed_meta.csv', index=False)
```
